```python
import boto3
import pandas as pd
from io import BytesIO

# AWS credentials
AWS_ACCESS_KEY_ID = 'xxxxx'
AWS_SECRET_ACCESS_KEY = 'yyy'
AWS_REGION = 'us-east-1'

# Initialize S3 resource
s3_resource = boto3.resource(
    service_name='s3',
    aws_access_key_id=AWS_ACCESS_KEY_ID,
    aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
    region_name=AWS_REGION
)

# Specify the bucket name
bucket_name = 'ecommerce-company-olist'

# List objects in the S3 bucket to verify the file key
bucket = s3_resource.Bucket(bucket_name)

print("Objects in S3 bucket:")
for obj in bucket.objects.all():
    print(obj.key)

# Specify the object key
file_key = 'new/'
```

```python
# Fetch the file from S3
try:
    obj = s3_resource.Object(bucket_name, file_key).get()
    df = pd.read_csv(BytesIO(obj['Body'].read()), index_col=0)
    # Display the DataFrame
    print(df.head())
except s3_resource.meta.client.exceptions.NoSuchKey:
    print(f"The specified key '{file_key}' does not exist in the bucket '{bucket_name}'.")

```
